```python
is_prime = np.ones((100,),dtype=bool)
print is_prime
```

```python
is_prime[0] = 0
is_prime[1] = 0
print is_prime
```

```python
N_max = int(np.sqrt(len(is_prime)))
for j in range(2, N_max):
    is_prime[2*j::j] = False
```

```python
def eratosthenes_sieve(maximum):
    is_prime = np.ones((maximum,),dtype=bool)
    is_prime[0] = 0
    is_prime[1] = 0
    
    for j in range(2, 10):
        is_prime[2*j::j] = False
    print np.nonzero(is_prime)
```

```python
eratosthenes_sieve(7)
```

```python
f = open('messy_data.dat', 'w')
import random
for i in range(100):
    for j in range(5):
        f.write(' ' * random.randint(0, 6))
        f.write('%0*.*g' % (random.randint(8, 12),
                            random.randint(5, 10),
                            100 * random.random()))
        if j != 4:
            f.write(',')
    f.write('\n')
f.close()
```

```python
!head -4 messy_data.dat
```

```python
a = open('messy_data.dat')
for line in a:
    for item in line.split(','):
        print item
        print '    '
    print '\n'
```

```python
fout = open('clean_data.dat','w')
for line in open('messy_data.dat'):
    for item in line.split(','):
        fout.write('%8.5f'%float(item))
        fout.write('    ')
    fout.write('\n')
    
fout.close()
        
```
