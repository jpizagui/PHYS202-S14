(a) See what happens to the period of the pendulum as you change the initial angle of release. Plot θ vs t for 5 different initial angles, θ0 , spanning 0 to π. (i.e. 0, π/4 , π/2 , 3π/4 , and π ). Put them all on the same plot. What happens to the period of the pendulum as the initial angle increases from π/4 to 3π/4? What happens to the trajectory when the initial angle is 0 and π ?


```python
def deriv(theta_current,t):  # return derivatives of the array theta
    g = 9.8 #m/s^2
    ell = 0.5 #m
    theta = theta_current[0]
    omega = theta_current[1]
    dtheta = omega
    domega = -(g/ell)*np.sin(theta)
    return np.array([ dtheta, domega ])

from scipy.integrate import odeint

t = np.arange(0.,10.+0.01,0.01)
th0 = 20.*pi/180 # 20 degrees, but must use radians
omega0 = 0. #starting from rest, omega = 0
theta_init = np.array([th0, omega0]) # initial values, th0 from before = 20 degrees, expressed in radians

#pass the function deriv to the solver with initial values and integration variable
theta = odeint(deriv,theta_init,t) 


```

```python
plt.figure(figsize=(8, 6), dpi=80)
for th0 in [0,45.*pi/180,90.*pi/180,135.*pi/180,180.*pi/180]:
    
    
    theta_init = np.array([th0, omega0]) # initial values
    theta = odeint(deriv,theta_init,t)
    if th0 == 0:
        m = "0 Degrees"
    if th0 == 45.*pi/180:
        m = "45 Degrees"
    if th0 == 90.*pi/180:
        m = "90 Degrees"
    if th0 == 135.*pi/180:
        m = "135 Degrees"
    if th0 == 180.*pi/180:
        m = "180 Degrees"
        
    plt.plot(t,theta[:,0],label=m)
plt.xlabel(r'$t (seconds)$',fontsize=20)
plt.ylabel(r'$\theta (rad)$',fontsize=20)
plt.grid()
plt.legend(loc='best')
plt.show()
```

(a) Use Euler's method (not odeint) to solve the following first-order differential equation with given initial condition y(0) = 0.

dydx=x+2y

Find the value of y when x = 1, y(1), using two different step sizes: 0.25 then 0.02.


```python
def EulerMethod(step):
    xmax = 1 # maximum x
    dx = step # time step
    nsteps = int(xmax/dx)+1
    x = np.zeros(nsteps)
    yprime = np.zeros(nsteps)
    y = np.zeros(nsteps)
 
    # set initial conditions, redundant in this example
    x[0] = 0.
    yprime[0] = 0.

    for i in range(len(x)-1):
        x[i+1] = x[i] + dx
        yprime[i+1] = x[i] + 2*y[i]
        y[i+1] = y[i] + dx * (x[i]+2*y[i])
        
    return x,y,yprime

```

```python
x,y,yprime = EulerMethod(.25)
print "The Value of Y When X = 1 is (Step Size = .25):", y[-1]
print y
x,y,yprime = EulerMethod(.02)
print "The Value of Y When X = 1 is (Step Size = .02):", y[-1]
```



(c) Now solve the same problem using Euler-Cromer and the Midpoint method for the same pair of step sizes and plot the results along with the exact solution. Then also plot the difference between the approximate numerical solutions and the exact solution. Which combination of method and step-size gives the best result?


```python
def EulerCromerMethod(step):
    xmax = 1 # maximum x
    dx = step # time step
    nsteps = int(xmax/dx)+1
    x = np.zeros(nsteps)
    yprime = np.zeros(nsteps)
    y = np.zeros(nsteps)
 
    # set initial conditions, redundant in this example
    x[0] = 0.
    yprime[0] = 0.

    for i in range(len(x)-1):
        x[i+1] = x[i] + dx
        yprime[i+1] = x[i] + 2*y[i]
        y[i+1] = y[i] + dx * (x[i+1]+2*y[i])
        
    return x,y,yprime

def MidpointMethod(step):
    xmax = 1 # maximum x
    dx = step # time step
    nsteps = int(xmax/dx)+1
    x = np.zeros(nsteps)
    yprime = np.zeros(nsteps)
    y = np.zeros(nsteps)
 
    # set initial conditions, redundant in this example
    x[0] = 0.
    yprime[0] = 0.

    for i in range(len(x)-1):
        x[i+1] = x[i] + dx
        yprime[i+1] = x[i] + 2*y[i]
        y[i+1] = y[i] + dx * ((1./2)*(x[i+1]+x[i])+2*y[i])
        
    return x,y,yprime
```

```python
x,y,yprime = EulerCromerMethod(.25)
yex=0.25*np.e**(2*x)-0.5*x-0.25
Error1 = abs(y-yex)
plt.plot(x,y,label="Step Size of .25")
x,y,yprime = EulerCromerMethod(.02)
yex=0.25*np.e**(2*x)-0.5*x-0.25
Error2 = abs(y-yex)
plt.plot(x,y,label="Step Size of .02")
yex=0.25*np.e**(2*x)-0.5*x-0.25
plt.plot(x,yex,label="Exact")
plt.legend(loc="best")
plt.xlim(0,1)
plt.show()
```

```python
x,y,yprime = MidpointMethod(.25)
plt.plot(x,y,label="Step Size of .25")
yex=0.25*np.e**(2*x)-0.5*x-0.25
Error3 = abs(y-yex)

x,y,yprime = MidpointMethod(.02)
plt.plot(x,y,label="Step Size of .02")
yex=0.25*np.e**(2*x)-0.5*x-0.25
Error4 = abs(y-yex)

y=0.25*np.e**(2*x)-0.5*x-0.25

plt.plot(x,y,label="Exact")
plt.legend(loc="best")
plt.xlim(0,1)
plt.show()
```

```python
plt.figure(figsize=(8, 6), dpi=80)
x,y,yprime = EulerCromerMethod(.25)
plt.plot(x,Error1,label="EulerCromer Method,step=.25")
x,y,yprime = EulerCromerMethod(.02)
plt.plot(x,Error2,label="EulerCromer Method,step=.02")
x,y,yprime = MidpointMethod(.25)
plt.plot(x,Error3,label="Midpoint Method,step=.25")
x,y,yprime = MidpointMethod(.02)
plt.plot(x,Error4,label="Midpoint Method,step=.02")
plt.xlim(0,1.0)
plt.xlabel("X",fontsize=20)
plt.ylabel("Error Of Methods",fontsize=20)
plt.legend(loc="best")
```

### It's clear that the best result comes from using the Euler-Cromer Method with a step of .02, Followed by the Midpoint Method with step .02

(a) Create a derivative function to solve this system of equations and use odeint to solve it for a projectile fired at an angle of 50 degrees from the horizontal at an initial speed of 30 m/s. Track the projectile's motion over the time interval of 0 to 200 seconds and let (x0,y0) = (0,0). You'll have to calculate (vx0,vy0) from the information provided. Let vter=44 m/s. Plot the trajectory of the projectile (y vs. x) to see how far it can travel in the x direction before landing.

```python
def PPderiv(y,t, Vterr=44):
    
    g = 9.8
    m = 1 #kg
    Vx = y[0]
    Vy = y[1]
    x = y[2]
    y = y[3]
    Vterr = 44
    
    dVxdt = (-g)*((Vx/(Vterr)**2))*np.sqrt((Vx)**2+(Vy)**2)
    dVydt = (-g)*(1+((Vy/(Vterr)**2)*np.sqrt((Vx)**2+(Vy)**2)))
    
    return np.array([ dVxdt, dVydt, Vx, Vy ])
```

```python
t = np.arange(0.,200.+0.1,0.1)

#initial conditions
Vx0 = 19.28
Vy0 = 22.98
x0 = 0
y0 = 0

#pass the function deriv to the solver with initial values and integration variable
result = odeint(PPderiv,np.array([Vx0,Vy0,x0,y0]),t)


X = result[:,2]
Y = result[:,3]

plt.plot(X,Y,label="Trajectory")
plt.xlabel("X")
plt.ylabel("Y")
plt.xlim(0,100)
plt.ylim(0,100)
plt.legend(loc="best")
plt.show()
```
